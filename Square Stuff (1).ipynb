```python
#Everything Together  

from sympy.abc import x, y 
import sympy as sp
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np

eig = int(input("Enter the number of unique eigenvalues that you want: "))
L = int(input("Enter length (Bohrs): "))
mass = int(input("Enter mass of particle (atomic units): "))

print ("");
m = 1;
n = 1;
#Here, (m,n) represent quantum numbers.

def H(m,n):
    B = np.sqrt(2/L)*sp.sin(m*sp.pi*x)
    C = np.sqrt(2/L)*sp.sin(n*sp.pi*y)
    D = B*C
    Htx = sp.diff(D,x,x) 
    Hty = sp.diff(D,y,y)
    Ht = -0.5*(Htx + Hty)
    Ht = Ht / mass
    Ht = sp.integrate(D*Ht, (x,0,L))
    Htt = sp.integrate(Ht, (y,0,L))
    Htt = round(Htt,3)
    return Htt

def Q(m,n):
    B = np.sqrt(2/L)*sp.sin(m*sp.pi*x)
    C = np.sqrt(2/L)*sp.sin(n*sp.pi*y)
    D = B*C
    Htx = sp.diff(D,x,x) 
    Hty = sp.diff(D,y,y)
    Ht = -0.5*(Htx + Hty)
    Ht = Ht / mass
    Ht = sp.integrate(D*Ht, (x,0,L))
    Ht = sp.integrate(Ht, (y,0,L))
    Htt = Ht,(m,n)
    return Htt

def M(m,n):
    return m

def N(m,n):
    return n 



results = [H(m, n) for m in range(1, 20) for n in range(1,20)]
resultsqn = [Q(m,n) for m in range(1,20) for n in range(1,20)]
resultsm = [M(m,n) for m in range(1,20) for n in range(1,20)]
resultsn = [N(m,n) for m in range(1,20) for n in range(1,20)]

# results = [];
# while m <= 100:  
#     while n <= 100:
#             results.append(H(m,n))
#             n = n + 1 
#     m = m + 1
#     n = 1


results, resultsqn, resultsm, resultsn = (list(t) for t in zip(*sorted(zip(results, resultsqn, resultsm, resultsn))))

eigcounter = 1
index = 1
deg = True

def P(m,n):
    def f(x, y):
        B = np.sqrt(2/L)*np.sin(m*np.pi*x) # Replace '1' with 'm' later
        C = np.sqrt(2/L)*np.sin(n*np.pi*y) # Replace '1' with 'n' later
        D = B*C
        return D

    x = np.linspace(0, 1, 50)
    y = np.linspace(0, 1, 50)

    X, Y = np.meshgrid(x, y)
    Z = f(X, Y)
    
    plt.figure(index)
    plt.contourf(X, Y, Z, 10, cmap = 'RdBu')
    plt.colorbar()
    if deg == True:
        plt.title(str((m,n))+str(' Degenerate'))
    else:
        plt.title((m,n))
        
    plt.xlabel('X')
    plt.ylabel('Y')

while eigcounter <= eig:
    if (results[index-1] == results[index]) and (results[index-1] != results[index-2]):
            if eigcounter == eig:
                print(eigcounter,')','Degeneracy', resultsqn[index-1])
                print('    Degeneracy', resultsqn[index],'\n')
                m = resultsm[index-1]
                n = resultsn[index-1]
                deg = True
                P(m,n)
                eigcounter = eigcounter + 1
                
            else:
                print(eigcounter,')','Degeneracy', resultsqn[index-1])
                m = resultsm[index-1]
                n = resultsn[index-1]
                deg = True
                P(m,n)
                index = index + 1 
                eigcounter = eigcounter + 1
    
            
    elif (results[index-1] != results[index]) and (results[index-1] != results[index-2]) :
            print (eigcounter,')',resultsqn[index-1],'\n')
            m = resultsm[index-1]
            n = resultsn[index-1]
            deg = False
            P(m,n)
            index = index + 1
            eigcounter = eigcounter + 1
            
            
    elif results[index-1] == results[index-2] and index < len(results):
    
            print ('    Degeneracy', resultsqn[index-1],'\n')
            m = resultsm[index-1]
            n = resultsn[index-1]
            deg = True
            P(m,n)
            index = index + 1
      

if (results[index-1] == results[index]) and (results[index-1] != results[index-2]) and (eigcounter == eig + 1):
        index = index + 1
        m = resultsm[index-1]
        n = resultsn[index-1]
        deg = True
        P(m,n)
            
print (resultsqn)
```
